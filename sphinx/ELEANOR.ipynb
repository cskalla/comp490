```python
if not os.path.exists('Voting_Precincts.geojson'):
    url = 'https://github.com/cskalla/comp490/blob/main/sphinx/Voting_Precincts.geojson'
    r = requests.get(url)
    with open('Voting_Precincts.geojson', 'w') as f:
        f.write(r.content.decode("utf-8"))

#load in geojson data       
with open('Voting_Precincts.geojson', 'r') as f:
    data = json.load(f) 
data
print(data['features']['PRECINCTID'])
```
